```python
print("min: ",df['Score'].min())
print("Q1: ",df['Score'].quantile(.25))
print("Q2: ",df['Score'].quantile(.5))
print("Q3: ",df['Score'].quantile(.75))
print("Max: ",df['Score'].max())
```

```python
# IQR = Q3 - Q1

Q1=df['Score'].quantile(.25)
Q3=df['Score'].quantile(.75)
```

```python
IQR = Q3-Q1

upperbound = Q3+1.5*IQR
lowerbound= Q1-1.5*IQR

print("IQR: ",IQR)
print("lowerbound: ", lowerbound)
print("upperbound: ", upperbound)
```
